```python
import numpy as np
import cv2
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from keras.models import load_model
from keras.utils import load_img, img_to_array
```

## Load Model 1

```python
model = load_model('model/model_224/mobilenet-244-rgb.h5')
model.summary()
```

## Making Prediction

```python
path = 'testing/test_bopeng/test_b_1.jpg'

img_pred = load_img(path, target_size=(224, 224))
imgplot = plt.imshow(img_pred)
x = img_to_array(img_pred)
x = np.expand_dims(x, axis=0)

images = np.vstack([x])
result = model.predict(images, batch_size=10)

if result[0,0]!=0:
    pred_labels = 'Bopeng'
elif result[0,1]!=0:
    pred_labels = 'Jerawat'
else:
    pred_labels = 'Komedo'

print(pred_labels)
```
